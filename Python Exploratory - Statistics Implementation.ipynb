# **2. Komparasi Simpangan Baku**

Pada bagian ini kalian akan mengamati nilai penyebaran data.

Simpangan baku (standard deviation) adalah ukuran statistik yang digunakan untuk mengukur sebaran atau dispersi data dalam sebuah sampel atau populasi. Simpangan baku menggambarkan sejauh mana nilai-nilai data tersebar di sekitar rata-rata (mean) dalam suatu distribusi.

\begin{equation}
\sigma = \sqrt{\frac{\sum{(x - \mu)^2}}{N}}
\end{equation}

* œÉ adalah simpangan baku.
* ùíô adalah nilai data dalam sampel atau populasi.
* Œº adalah rata-rata (mean) dari data.
* ‚àë adalah simbol sigma yang menunjukkan penjumlahan.
* ùëµ adalah jumlah total nilai dalam sampel atau populasi.
<br><br>

## üìñ **Soal**

Karena ketidakpastian situasi global, memaksa DQTech untuk melakukan efisiensi. Dari beberapa produk yang DQTech jual, DQTech harus mengeliminasi 1 produk. Data penjualan pada periode waktu tertentu disajikan pada list berikut :

Penjualan produk A (secara kuantitas tiap bulannya)
```
penjualan_produk_A = [55, 49, 48, 50, 52, 47, 51, 50, 53, 48, 50, 49, 51, 46, 50]
```

Penjualan produk B (secara kuantitas tiap bulannya)
```
penjualan_produk_B = [80, 20, 45, 65, 30, 90, 10, 110, 60, 55, 75, 35, 48, 50, 25]
```

Coba hitung simpangan baku dari `penjualan_produk_A` dan `penjualan_produk_B` dan bantu DQTech menentukan mana produk yang sebaiknya dipertahankan? Mengapa?

````python
penjualan_produk_A = [55, 49, 48, 50, 52, 47, 51, 50, 53, 48, 50, 49, 51, 46, 50]
penjualan_produk_B = [80, 20, 45, 65, 30, 90, 10, 110, 60, 55, 75, 35, 48, 50, 25]

# MENGHITUNG RATA-RATA PENJUALAN PRODUK A & B
avg_penjualan_prd_A = sum(penjualan_produk_A) / len(penjualan_produk_A)
avg_penjualan_prd_B = sum(penjualan_produk_B) / len(penjualan_produk_B)

print(f'Rata-rata penjualan untuk produk A adalah {avg_penjualan_prd_A:.2f}')
print(f'Rata-rata penjualan untuk produk B adalah {avg_penjualan_prd_B:.2f}')
````

````python
# FUNGSI SIMPANGAN BAKU
def calculate_std_deviation(data):
  n = len(data)
  avg = sum(data) / n
  variance = sum((x - avg) ** 2 for x in data) / n
  s_deviation = variance ** 0.5

  return s_deviation

# MENGHITUNG SIMPANGAN BAKU
std_dev_prd_A = calculate_std_deviation(penjualan_produk_A)
std_dev_prd_B = calculate_std_deviation(penjualan_produk_B)

# MENAMPILKAN HASIL
print(f'Standard Deviation untuk produk A adalah {std_dev_prd_A:.2f}')
print(f'Standard Deviation untuk produk B adalah {std_dev_prd_B:.2f}')
````

````python
# MENAMPILKAN VISUALISASI
import matplotlib.pyplot as plt

# MONTH TIME SERIES UNTUK VISUALISASI
month = list(range(1, len(penjualan_produk_A) + 1))

# VISUALISASI
plt.figure(figsize = (12, 6))
plt.plot(month, penjualan_produk_A, color = 'green', label = 'Produk A', marker = 'o', linestyle = '-')
plt.plot(month, penjualan_produk_B, color = 'red', label = 'Produk B', marker = 's', linestyle = '--')

# TAMBAH RATA-RATA KE GRAFIK
plt.axhline(y = avg_penjualan_prd_A, color = 'green', linestyle = ':', label = f"Rata-rata A: {avg_penjualan_prd_A:.2f}")
plt.axhline(y = avg_penjualan_prd_B, color = 'red', linestyle = ':', label = f"Rata-rata B: {avg_penjualan_prd_B:.2f}")

# TAMBAH JUDUL DAN LABEL
plt.title("Penjualan Produk A dan B per Bulan", fontsize=14)
plt.xlabel("Bulan", fontsize=12)
plt.ylabel("Jumlah Penjualan", fontsize=12)
plt.xticks(month)
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()

# MENAMPILKAN VISUALISASI
plt.show()
````
